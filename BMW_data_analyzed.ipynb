```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
bmw_stock = pd.read_csv('BMW_stock_data.csv')
bmw_data = pd.read_csv('BMW_data.csv')

# Display the first few rows of the datasets
bmw_data.head(10)
```

```python
bmw_stock.head(10)
```

```python
# shape of the datasets
print("Shape of BMW Stock Data:", bmw_stock.shape)
print("Shape of BMW Data:", bmw_data.shape)
```

```python
# Check for missing values, data types, and summary statistics
print("BMW Stock Data Info:")
print(bmw_stock.info())
print("\nBMW Stock Data Summary Statistics:")
print(bmw_stock.describe())
```

```python
print("\nBMW Data Info:")
print(bmw_data.info())
```

```python
print("\nBMW Data summary statistics:")
print(bmw_data.describe())
```

```python
#change date column to datetime format
bmw_stock['Date'] = pd.to_datetime(bmw_stock['Date'])
```

```python
# display date colimn after conversion
print(bmw_stock['Date'].head(10))
```

```python
# add a new column for year
bmw_stock['Year'] = bmw_stock['Date'].dt.year
```

```python
bmw_stock.columns
```

```python
# compare the distribution of stock prices over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=bmw_stock, x='Year', y='Close', label='Closing Price')
plt.title('BMW Stock Closing Price Over Years')
plt.xlabel('Year')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# correlation matrix for BMW stock data
plt.figure(figsize=(10, 8))
sns.heatmap(bmw_stock.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix for BMW Stock Data')
plt.show()
```

```python
# group by year and calculate the average closing price
avg_closing_price = bmw_stock.groupby('Year')['Close'].mean().reset_index()
# plot the average closing price over years
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_closing_price, x='Year', y='Close', marker='o')
plt.title('Average BMW Stock Closing Price Over Years')
plt.xlabel('Year')
plt.ylabel('Average Closing Price')
plt.xticks(rotation=45)
sns.regplot(data=avg_closing_price, x='Year', y='Close', scatter=False, color='red')
plt.savefig('average_closing_price_over_years.png')
plt.show()
```

```python
# a barh plot to show the volime of stocks traded each year
# group by year and calculate the total volume
total_volume = bmw_stock.groupby('Year')['Volume'].sum().reset_index()
# plot the total volume over years
plt.figure(figsize=(12, 6))
sns.barplot(data=total_volume, x='Year', y='Volume', palette='viridis')
plt.title('Total Volume of BMW Stocks Traded Each Year')
plt.xlabel('Year')
plt.ylabel('Total Volume')
plt.xticks(rotation=45)
plt.savefig('total_volume_bmw.png')
plt.show()
```

```python
sns.relplot(data=bmw_stock, x='Open', y='Close', hue='Year', palette='viridis', alpha=0.5)
```

```python
#line chart to show relationship between the high and low stock prices over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=bmw_stock, x='Year', y='High', label='High Price')
sns.lineplot(data=bmw_stock, x='Year', y='Low', label='Low Price')
plt.title('BMW Stock High and Low Prices Over Years')
plt.xlabel('Year')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend()
plt.savefig('high_low_prices_over_years.png')
plt.show()
```

```python
bmw_data.columns
```

```python
# convert date column to datetime format
bmw_data['Date'] = pd.to_datetime(bmw_data['Date'])
```

```python
# generate a year column from the date column in bmw_data
bmw_data['Date'] = pd.to_datetime(bmw_data['Date'])
bmw_data['Year'] = bmw_data['Date'].dt.year
```

```python
# drop the date column from bmw_data
bmw_data.drop('Date', axis=1, inplace=True)
```

```python
bmw_data.head(10)
```

```python
bmw_data.tail(10)
```

```python
bmw_data['High'].mean()
```

```python
# occurence of high stock prices above 100 in the year 2025
high_price_2024 = bmw_data[(bmw_data['Year'] == 2024) & (bmw_data['High'] > 100)].shape[0]
print("Number of occurrences of high stock prices above 100 in 2024:", high_price_2024)
```

```python
# count the volume of stocks for the year 2024
volume_2024 = bmw_data[bmw_data['Year'] == 2024]['Volume'].sum()
print("Total Volume of BMW Stocks in 2024:", volume_2024)
```

```python
# closing price over years
plt.figure(figsize=(12, 6))
sns.lineplot(data=bmw_data, x='Year', y='Close', label='Closing Price')
plt.title('BMW Stock Closing Price Over Years')
plt.xlabel('Year')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.legend()
plt.savefig('closing_price_over_years.png')
# add a regression line to the plot
sns.regplot(data=bmw_data, x='Year', y='Close', scatter=False, color='red')
plt.show()
```
